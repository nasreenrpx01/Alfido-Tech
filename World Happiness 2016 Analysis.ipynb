```python
top_10_happiness = df.nlargest(10, 'Happiness Score')
top_10_happiness.plot(kind='bar', x='Country', y='Happiness Score', title='Top 10 Happiest Countries')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()
```

```python
df.isna()
```

```python
top_10_rank=df[df['Happiness Rank']<11]
top_10_rank
```

```python
region_trust_means = df.groupby('Region')['Trust (Government Corruption)'].mean()
region_trust_means.plot(kind='bar', xlabel='Region', ylabel='Average Trust Score', title='Trust by Region')
plt.show()
```

```python
max_trust_index = df["Trust (Government Corruption)"].idxmax()  
max_trust_row = df.loc[max_trust_index]
print(max_trust_row)
```

```python
generosity_by_region = df.groupby('Region')['Generosity'].sum()
generosity_by_region.plot(kind='pie', autopct='%.1f%%')
plt.title('Generosity by Region')
plt.show()
```

```python
df.boxplot(column='Health (Life Expectancy)', by='Region')
plt.title('Health Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Health (Life Expectancy)')
plt.xticks(rotation=45)
plt.show()
```

```python
df.plot(kind='scatter', x='Economy (GDP per Capita)', y='Happiness Score', title='Happiness vs. Economy')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(data=df, x='Region', y='Happiness Score')
plt.xticks(rotation=90)
plt.title('Happiness Score by Region')
plt.show()
```

```python
df['Freedom'].plot(kind='hist', bins=20, edgecolor='k', xlabel='Freedom Score', ylabel='Frequency', title='Distribution of Freedom Score')
plt.show()
```

```python
trust_by_region = df.groupby('Region')['Trust (Government Corruption)'].mean()
trust_by_region.plot(kind='pie', autopct='%1.1f%%', startangle=90, title='Trust by Region')
plt.axis('equal')
plt.show()
```

```python
economy_health = df[['Region', 'Economy (GDP per Capita)', 'Health (Life Expectancy)']]
economy_health.groupby('Region').sum().plot(kind='bar', stacked=True, title='Economy and Health Contribution by Region')
plt.ylabel('Sum')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Dystopia Residual'].plot(kind='hist', bins=15, edgecolor='k', xlabel='Dystopia Residual', ylabel='Frequency', title='Dystopia Residual Distribution')
plt.show()
```
