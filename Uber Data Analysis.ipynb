```python
df['START'] = df['START'].replace("Kar?chi", "Karachi")
df['STOP'] = df['STOP'].replace("Kar?chi", "Karachi")
```

```python
df["CATEGORY"].fillna("Not Mentioned",inplace=True)
```

```python
df['PURPOSE'].fillna("Not Mentioned", inplace=True)
```

```python
df['PURPOSE'].unique()
```

```python
Long_distance=df[df['MILES']>15]
Long_distance
```

```python
df['START_DATE'] = pd.to_datetime(df['START_DATE'], errors='coerce')
df['END_DATE'] = pd.to_datetime(df['END_DATE'], errors='coerce')
df.head()
```

```python
df['Hour'] = df['START_DATE'].dt.hour
df.head()
```

```python
Evening_Ride=df[df['Hour']>17]
Evening_Ride
```

```python
df['CATEGORY'].value_counts().plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
df.dropna(subset=['END_DATE'], inplace=True)
df.dropna(subset=['CATEGORY'], inplace=True)
df.dropna(subset=['START'], inplace=True)
df.dropna(subset=['STOP'], inplace=True)
```

```python
df.info()
```

```python
duplicates = df.duplicated()
print(duplicates.sum())
```

```python
purpose_counts = df['PURPOSE'].value_counts()
purpose_counts.plot(kind='bar', title='Trips by Purpose')
plt.xlabel('Purpose')
plt.ylabel('Count')
plt.show()
```

```python
df['MILES'].plot(kind='hist', bins=10, edgecolor='k')
plt.title('Miles Distribution')
plt.xlabel('Miles')
plt.ylabel('Frequency')
plt.show()
```

```python
hourly_miles = df.groupby('Hour')['MILES'].sum()
hourly_miles.plot(kind='line')
plt.title('Miles Driven Over Time')
plt.xlabel('Hour')
plt.ylabel('Total Miles')
plt.grid(True)
plt.show()
```

```python
df.boxplot(column='MILES', by='PURPOSE')
plt.title('Miles Distribution by Purpose')
plt.xlabel('Purpose')
plt.ylabel('Miles')
plt.xticks(rotation=45)
plt.show()
```

```python
df.plot(kind='scatter', x='Hour', y='MILES', s=50, alpha=0.5)
plt.title('Scatter Plot: Miles vs. Hour')
plt.xlabel('Hour')
plt.ylabel('Miles')
plt.show()
```

```python
df['START_DATE'] = pd.to_datetime(df['START_DATE'])
df.set_index('START_DATE')['MILES'].plot()
plt.title('Miles Driven Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Miles')
plt.grid(True)
plt.show()
```
